```python
plt.style.use('seaborn-v0_8-darkgrid')
x_cordinates = np.array([0,1,2])*2
plt.bar(x_cordinates,[10,20,30], color = 'orange',width = 0.5,label = "Current Price", tick_label =["platinum", "Gold", "Silver"])
plt.bar(x_cordinates+0.5,[2,16,10], width = 0.5,label ="Future Price",color ='r')
plt.title("Metal price comparsion")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,50)
plt.xlim(0,5)
plt.legend()
plt.show()
```

```python
print(titles, type(titles)) # it is a type of a pandas . series
```

```python
titles = list(df.get("Movie Name")) # to run loops
```

```python
print(titles)
```

```python
print(type(titles), len(titles))
```

```python
print(titles[:5])
```

```python
print(titles[0][0: 4])
```

```python
freq_titles = {}
```

```python
for title in titles:
    length = len(title)
    if freq_titles.get(length) is None:
        freq_titles[length] = 1 # dictionary of new movie with uniquer name
    else:
        freq_titles[length] += 1 # 
    
```

```python
freq_titles
```

```python
print(freq_titles)
```

```python
X = np.array(list(freq_titles.keys()))
```

```python
X
```

```python
Y = np.array(list(freq_titles.values()))
```

```python
Y
```

```python
print(X, Y)
```

```python
plt.scatter(X, Y, color = 'r')
plt.xlabel("Lenght of movie title")
plt.ylabel("frequency count")
plt.title("Movie DV")
plt.grid()
plt.show()
```
