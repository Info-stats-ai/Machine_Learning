  6) Summation(i to m )|y<i> (actual ) - y<i> (predicted)|, m is the total number of exaples in training data

  9) Since y^(i)(predicted)  = hypothesis(x(i)) = theta-1 *x(i) + theta-0

  11) update theta so that j(theta reduces)

### LR - Implementating Gradeint Descent

```python
def hypothesis(x, theta):
    y_prediction = 0
   # theta is a vector with two values , theta nod and theta 1
    y_prediction = theta[0] + theta[1]*x
    return y_prediction 
```

```python
print(np.shape(Y))
```

```python
def gradient(X, Y, theta):
    m = X.shape[0]
    grad = np.zeros((2,))
    for i in range(m):
       
        y_prediction = hypothesis(X[i], theta)
        
        grad[0] += (y_prediction - Y[i]) # summation of gradient theta 0
        grad[1] += (y_prediction - Y[i])*X[i]
    return grad / m #formula for j(0) = 1/m summation (o to m ) y_- y
    
```

```python
X.shape[0]
```

```python
def error(X, Y, theta):
    m = X.shape[0]
    total_error = 0.0
    for i in range(m):
        y_prediction = hypothesis(X[i], theta)
        total_error += (y_prediction - Y[i])**2 # mean squared error
    return total_error / m # mean error
        
```

```python
def gradientDescent(X, Y, lr = 0.1, max_steps = 100):
    theta = np.zeros((2,))
    error_list = []
    for i in range(max_steps):
       #compute the gradient
        grad = gradient(X, Y, theta)
        e = error(X, Y, theta)
        error_list.append(e)
        #update theta
        theta[0] = theta[0] - lr*grad[0]
        theta[1] = theta[1] - lr*grad[1]
    return theta, error_list
        
```

### Computing Score
- R2(R-squared) or Coefficient of determination
-  R2 score =1 - (summation (1 to m) {(yactual - ypredicted)**2})/ summation (1 to m){(yactual - yavg)**2}.
-  Maximum value R2 score is can be 1.
-  Minimum value can be negative also.
-  R2 score  in best score = 1
-  R2 score in worst case --->>>> ypred = yavg. = 0
-  accuracy = 100*R2-score

```python
def r2_score(Y_twentypercent_for_test ,y_test_predicted) :
    numerator = np.sum(Y_twentypercent_for_test - y_test_predicted)**2
    denominator = np.sum((Y_twentypercent_for_test - Y_twentypercent_for_test.mean())**2)
    score = (1 - numerator/denominator)
    return score*100
```
