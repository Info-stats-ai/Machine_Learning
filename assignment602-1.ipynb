- Each randfom generator produces different set values
- To solve this we will use rng.
- we use np.random.default_rng() when- ever we perform calculations involving random quantities within numpy. In principle, this should enable the reader to exactly reproduce the stated results.

```python
rng = np.random.default_rng(1) 
print(rng.normal(loc = 100, scale=2, size=(1,2))) 
rng2 = np.random.default_rng(1) 
print(rng2.normal(loc = 100, scale=2, size=(1,2)))
```
